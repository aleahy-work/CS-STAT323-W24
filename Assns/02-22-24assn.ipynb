Consider the problem of approximating the function given below by a neural network.  (This is very similar to the regression problem given in the previous assignment--a key should be available by now if you had difficulty with this.)  

```python
import numpy as np
import matplotlib.pyplot as plt
tempX = np.random.uniform(-1,1,4000)
tempY = np.random.uniform(-1,1,4000)

myX = np.array([tempX, tempY]).T

myX[:5]

def myfunc(X):
    
    return  np.sin((X[:,0]**2 + X[:,1]**2)*2)

myy = myfunc(myX)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(myX[:,0], myX[:,1], myy, s= 1)
plt.show()
```
